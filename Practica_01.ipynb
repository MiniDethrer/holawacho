**1. Calcular el área de un triángulo**

```python
print("--- Calculadora de Área de Triángulo ---")
base = float(input("Introduce la base del triángulo: "))
altura = float(input("Introduce la altura del triángulo: "))
area = (base * altura) / 2
print(f"El área del triángulo es: {area}")
```

**2. Convertir Celsius a Fahrenheit**

```python
print("\n--- Conversor de Temperatura ---")
celsius = float(input("Introduce la temperatura en grados Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C equivale a {fahrenheit:.2f}°F")
```

**3. Indicar el número mayor**

```python
print("\n--- Comparador de Tres Números ---")
num1 = float(input("Introduce el primer número: "))
num2 = float(input("Introduce el segundo número: "))
num3 = float(input("Introduce el tercer número: "))
if num1 >= num2 and num1 >= num3:
    mayor = num1
elif num2 >= num1 and num2 >= num3:
    mayor = num2
else:
    mayor = num3
print(f"El número mayor de los tres es: {mayor}")
```

**4. Calcular descuento por compra**

```python
print("\n--- Calculadora de Descuentos ---")
monto_total = float(input("Introduce el monto total de la compra en S/: "))
descuento = 0.0
if monto_total > 500:
    descuento = monto_total * 0.20
    print("Se aplicó un descuento del 20%.")
elif monto_total >= 200:
    descuento = monto_total * 0.10
    print("Se aplicó un descuento del 10%.")
else:
    print("No se aplica ningún descuento.")
monto_final = monto_total - descuento
print(f"Monto original: S/{monto_total:.2f}")
print(f"Descuento: S/{descuento:.2f}")
print(f"Monto final a pagar: S/{monto_final:.2f}")
```

**5. Suma de números pares en un rango**

```python
print("\n--- Suma de Pares en un Rango ---")
inicio = int(input("Introduce el número de inicio del rango: "))
fin = int(input("Introduce el número de fin del rango: "))
suma_pares = 0
for numero in range(inicio, fin + 1):
    if numero % 2 == 0:
        suma_pares += numero
print(f"La suma de todos los números pares entre {inicio} y {fin} es: {suma_pares}")
```
